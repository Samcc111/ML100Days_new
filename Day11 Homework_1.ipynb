### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
```


````python
# 記得先 Import 正確的套件

import numpy as np
import pandas as pd
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

````python
# cat_meanage = np.nanmean(df[df['animal']=='cat']['age'])
# snake_meanage = np.nanmean(df[df['animal']=='snake']['age'])
# dog_meanage = np.nanmean(df[df['animal']=='dog']['age'])

cat_meanage = (df[df['animal']=='cat']['age']).mean()
snake_meanage = (df[df['animal']=='snake']['age']).mean()
dog_meanage = (df[df['animal']=='dog']['age']).mean()
print('貓的平均年齡',cat_meanage)
print('蛇的平均年齡',snake_meanage)
print('狗的平均年齡',dog_meanage)
````

````python
aa = df.sort_values(by=['age'])
bb = aa.sort_values(by=['visits'], ascending=False)
bb
````

````python
df01= df.copy()
df01['priority'] = df01['priority'].str.replace('yes','True')
df01['priority'] = df01['priority'].str.replace('no','False')
df01
````

### 2. 一個包含兩個欄位的 DataFrame，將每個數字減去

* 1) 該欄位的平均數
* 2) 該筆資料平均數

````python
df = pd.DataFrame(np.random.random(size=(5, 3)),columns=['A','B','C'])
df
````

````python
df01 = df.copy()
df01 = df01 -df01.mean()
# df01['A'] =df01['A'] - df01.A.mean()
# df01['B'] =df01['B'] - df01.B.mean()
# df01['C'] =df01['C'] - df01.C.mean()
df01
````

````python
df02 = df.copy()
mean_df = df02.mean(axis =1)
````

````python
df02.iloc[0] = df02.iloc[0] - mean_df.iloc[0]
df02.iloc[1] = df02.iloc[1] - mean_df.iloc[1]
df02.iloc[2] = df02.iloc[2] - mean_df.iloc[2]
df02.iloc[3] = df02.iloc[3] - mean_df.iloc[3]
df02.iloc[4] = df02.iloc[4] - mean_df.iloc[4]
df02
````
